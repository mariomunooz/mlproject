```python
def set_palette_and_order(df, column):
    """
    Set a color palette and order for the specified column in the DataFrame.

    Parameters:
    df (DataFrame): The input DataFrame.
    column (str): The column name for which to set the palette and order.

    Returns:
    tuple: (palette, order) where palette is a dictionary of colors and order is a list of unique categories.
    """
    # Get unique categories from the specified column
    unique_categories = df[column].unique()
    
    # Create a color palette based on the number of unique categories
    palette = sns.color_palette("tab10", n_colors=len(unique_categories))
    
    # Create a mapping of categories to colors
    category_palette = dict(zip(unique_categories, palette))
    
    # Sort categories for consistent ordering
    order = sorted(unique_categories)  # or specify your own order

    return category_palette, order
```

```python
# Set the palette and order for the 'lunch' column
lunch_palette, lunch_order = set_palette_and_order(df, 'lunch')

# Create subplots
plt.subplots(1, 3, figsize=(25, 6))
plt.subplot(141)
sns.histplot(data=df, x='average', kde=True, hue='lunch', palette=lunch_palette, hue_order=lunch_order)
plt.title('Average Scores by Lunch (All)')

plt.subplot(142)
sns.histplot(data=df[df.gender == 'female'], x='average', kde=True, hue='lunch', palette=lunch_palette, hue_order=lunch_order)
plt.title('Average Scores by Lunch (Female)')

plt.subplot(143)
sns.histplot(data=df[df.gender == 'male'], x='average', kde=True, hue='lunch', palette=lunch_palette, hue_order=lunch_order)
plt.title('Average Scores by Lunch (Male)')

plt.show()
```

#####  Insights
- Standard lunch helps perform well in exams.
- Standard lunch helps perform well in exams be it a male or a female.

```python
# Set the palette and order for the 'parental_level_of_education' column
education_palette, education_order = set_palette_and_order(df, 'parental_level_of_education')

# Create subplots
plt.subplots(1, 3, figsize=(25, 6))

plt.subplot(141)
sns.histplot(data=df, x='average', kde=True, hue='parental_level_of_education', palette=education_palette, hue_order=education_order)
plt.title('Average by Parental Level of Education (All)')

plt.subplot(142)
sns.histplot(data=df[df.gender == 'male'], x='average', kde=True, hue='parental_level_of_education', palette=education_palette, hue_order=education_order)
plt.title('Average by Parental Level of Education (Male)')

plt.subplot(143)
sns.histplot(data=df[df.gender == 'female'], x='average', kde=True, hue='parental_level_of_education', palette=education_palette, hue_order=education_order)
plt.title('Average by Parental Level of Education (Female)')

plt.show()
```

#### 4.3 Multivariate analysis using pieplot

```python
# Set the figure size and colormap
plt.rcParams['figure.figsize'] = (30, 12)
colors = plt.get_cmap('tab10').colors
textprops = {'fontsize': 20}

# Function to generate a pie chart
def plot_pie(data, title, color_indices):
    labels = data.index  # Get unique labels from index (value counts keys)
    plt.pie(data, colors=[colors[i] for i in color_indices], labels=labels, autopct='%1.2f%%', textprops=textprops)
    plt.title(title, fontsize=20)
    plt.axis('off')

# Plot all subplots
features = [
    ('gender', 'Gender', range(2)),
    ('race_ethnicity', 'Race/Ethnicity', range(5)),
    ('lunch', 'Lunch', range(2)),
    ('test_preparation_course', 'Test Course', range(2)),
    ('parental_level_of_education', 'Parental Education', range(6))
]


for i, (col, title, color_indices) in enumerate(features, 1):
    plt.subplot(2, 3, i)
    plot_pie(df[col].value_counts(), title, color_indices)

plt.tight_layout()
plt.show()
```

#####  Insights
- Number of Male and Female students is almost equal
- Number students are greatest in Group C
- Number of students who have standard lunch are greater
- Number of students who have not enrolled in any test preparation course is greater
- Number of students whose parental education is "Some College" is greater followed closely by "Associate's Degree"
